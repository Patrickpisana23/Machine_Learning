# Creación del Modelo KMeans

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=2)
```

```python
etiqueta_grupo = modelo.fit_predict(X_escala)
```

```python
etiqueta_grupo
```

```python
len(etiqueta_grupo)
```

```python
X['Grupo'] = etiqueta_grupo
```

# Elección del K adecuado

```python
ssd = []
for k in range(2,14):
    modelo = KMeans(n_clusters = k,random_state=101)
    modelo.fit(X_escala)
    ssd.append(modelo.inertia_)
```

```python
plt.plot(range(2,14),ssd[0:],'o--')
plt.xlabel("Valor de K")
plt.ylabel("Suma de distancias al cuadrado (ssd)")
```

```python
plt.plot(range(2,10),ssd[0:8],'o--')
plt.xlabel("Valor de K")
plt.ylabel("Suma de distancias al cuadrado (ssd)")
```

```python
plt.plot(range(2,15),ssd[0:13],'o--')
plt.xlabel("Valor de K")
plt.ylabel("Suma de distancias al cuadrado (ssd)")
```
