## Construcción del modelo de Regresión Logística


### División en conjunto de entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split
```

```python
X = datos.drop(['Pressure'],axis=1)
y = datos['Pressure']
```

# MODELO 1

### Para random_state=101

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.30,
                                                    random_state=101)
```

### Entrenamiento y Predicción

crear modelo

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
```

### Hiperparametros por defecto

```python
modelo.get_params()
```

### Entrenando el modelo

```python
modelo.fit(X_train,y_train)
```

### Predicción del modelo en el conjunto de pruebas

```python
prediccion =  modelo.predict(X_test)
```

```python
# 5 primeros valores verdaderos(reales)
```

```python
y_test[0:5]
```

```python
X_test[0:5]
```

```python
plt.scatter(y_test, prediccion)
```

```python
sns.distplot((y_test-prediccion),bins=50)
```

# Metricas de Evaluación de Medición

# MODELO 1

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, prediccion))
print('MSE:', metrics.mean_squared_error(y_test, prediccion))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediccion)))
```
