## Dividir en conjunto de entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split
```

```python
X = datos.drop('Cancer Present', axis=1)
y = datos['Cancer Present']
```

```python
X_test, X_train, y_test, y_train = train_test_split(X,
                                                   y,
                                                   test_size=0.3,
                                                   random_state=50)
```

Estandarizar datos


```python
from sklearn.preprocessing import StandardScaler
```

```python
escala = StandardScaler()
```

```python
escala.fit(X_train)
```

Transformamos de la entrada de entrenamiento y prueba a la escala

```python
X_train_escala = escala.transform(X_train)
```

```python
X_test_escala = escala.transform(X_test)
```

Crear el modelo KNN con K=1

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
modelo_knn = KNeighborsClassifier(n_neighbors=1)
```

Entrenamos el modelo

```python
modelo_knn.fit(X_train_escala, y_train)
```

Evaluamos el modelo


```python
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score
```

```python
y_pred = modelo_knn.predict(X_test_escala)
```

```python
plot_confusion_matrix(modelo_knn, X_test_escala,y_test, colorbar=False)
```

```python
print("Accuracy (exactitud) = ",accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

Metodo de ELBOW  para la eleccion del valor de K


```python
tasa_error = []

# Tomará algún tiempo
for i in range(1,61):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_escala,y_train)
    pred_test_i = knn.predict(X_test_escala)  
    tasa_error_i = 1-accuracy_score(y_test,pred_test_i)
    tasa_error.append(tasa_error_i)
```

```python
plt.figure(figsize=(8,6), dpi=150)
plt.plot(range(1,61),tasa_error,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10, label='Tasa Error de Prueba')
plt.title('Tasa de Error vs. Valor de K')
plt.xlabel('K')
plt.ylabel('Tasa de Error')
```

Crear el modelo KNN con K=19 y evaluarlo

```python
modelo_knn = KNeighborsClassifier(n_neighbors=28)
```

```python
modelo_knn.fit(X_train_escala, y_train)
```

```python
y_pred = modelo_knn.predict(X_test_escala)
```

```python
plot_confusion_matrix(modelo_knn, X_test_escala,y_test, colorbar=False)
```

```python
print("Accuracy (exactitud) = ",accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

### VALIDACION CRUZADA Y BUSQUEDA GRID PARA EL VALOR DE K

Creamos un Pipeline(Flujo) para encontra el valor de K

Objeto pipeline para establecer el flujo de trabajo de las operaciones
 1. Estandarizacion de datos 
 2. Crear el modelo con los datos estandarizados

Cuando se aplica estandarización dentro de un Pipeline ocurre lo siguiente

1. Los datos son divididos en datos de ENTRENAMIENTO y PRUEBA de acuerdo al parametro cv que se especifique en GridSearchCV.
2. La escala es ajustada (entrenada) con los datos de ENTRENAMIENTO
3. La escala transforma los datos de ENTRENAMIENTO
4. Los modelos se ajustan (entrenan) usando los datos de ENTRENAMIENTO transformados
5. La escala se usa para transformar los datos de PRUEBA
6. Los modelo entrenados predicen usando los datos de PRUEBA Transformar
 
    
    

```python
escala = StandardScaler()
```

```python
knn = KNeighborsClassifier()
```

```python
knn.get_params()
```

```python
knn.get_params().keys()
```

Se define las operaciones

```python
operaciones = [('Estandarizar', escala),('mod_knn', knn)]
```

Crear el Pipeline 

```python
from sklearn.pipeline import Pipeline
```

```python
flujo = Pipeline(operaciones)
```
